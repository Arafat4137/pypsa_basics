Additionally, 3 generators are added to bus D. The installed capacity is defined with the `p_nom` values and the marginal cost of producing 1 MWh (marginal_cost) is defined. The marginal cost is described as an optional attribute, but this must be defined in order to be able to run an optimal power flow.

```python
n.add(
    'Generator',
    name = ['Coal plant', 'Gas plant', 'Oil plant'],
    carrier = ['coal', 'gas', 'oil'],
    bus = 'D',
    p_nom = [100, 50, 20],
    marginal_cost = [10, 20, 100]
)
```

You can define custom function like the one below to plot the network with bus sizes and colors adjusted based on load set points and generation installed capacity.

```python
def plot_network():
    gen = n.generators.assign(g=n.generators.p_nom).groupby(["bus", "carrier"]).g.sum()
    load = n.loads.assign(l=n.loads.p_set).groupby(['bus', 'carrier']).l.sum()
    bus_data = pd.concat([gen, load])
    plt.figure(figsize = (20,10))
    n.plot(geomap = False, bus_sizes = bus_data/200, margin = 0.12,
           bus_colors = {'coal':'grey', 'gas':'orange','oil':'brown', 'load':'green'})
    plt.show()
```

```python
plot_network()
```

```python
def plot_network_flow():
    gen = n.generators.assign(g=n.generators_t.p.mean()).groupby(["bus", "carrier"]).g.sum()
    load = n.loads.assign(l=n.loads_t.p.mean()).groupby(['bus', 'carrier']).l.sum()
    bus_data = pd.concat([gen, load,])
    plt.figure(figsize = (20,10))
    n.plot(geomap = False, flow = "mean", bus_sizes = bus_data/200, margin = 0.12,
           bus_colors = {'coal':'grey', 'gas':'orange','oil':'brown', 'load':'green'})
    plt.show()
```
